```python
import sklearn
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
```

```python
clean_data = pd.read_pickle("cleaned_data.pkl")
```

```python
clean_data["expert"] = pd.Categorical(
    clean_data["expert"],
    categories=sorted(clean_data["expert"].unique(), key=lambda x: int(x.split('_')[1])),
    ordered=True
)

f, (ax1, ax2) = plt.subplots(2, 1, figsize=(13,6))

palette = ['yellow', 'black', 'white']

# plt.figure(figsize=(10, 4))
ax1 = sn.boxplot(x="expert", hue="definition", y="F1", data=clean_data.reset_index(), ax=ax1, palette=palette)
ax1.set(xticklabels=[])
ax1.axhline(clean_data["F1"].mean(), ls="--", color="blue", label="Mean F1 over all")
ax1 = sn.lineplot(data=clean_data.groupby(["expert"])["F1"].mean(), color="black", marker="s", label="Mean F1 per Expert")
legend_labels, _= ax1.get_legend_handles_labels()
ax1.legend(ncol=4)
# ax1.tick_params(bottom=False)
# ax1.set_xlabel(None)
# ax.set_xticklabels([])
ax1.ylim=(0,1)


ax2 = sn.boxplot(x="expert", hue="dataset", y="F1", data=clean_data.reset_index(), ax=ax2)
ax2.axhline(clean_data["F1"].mean(), ls="--", color="blue", label="Mean F1 over all")
ax2 = sn.lineplot(data=clean_data.groupby(["expert"])["F1"].mean(), color="black", marker="s", label="Mean F1 per Expert")
legend_labels, _= ax2.get_legend_handles_labels()
ax2.legend(ncol=4)
ax2.set(ylim=(0,1))
```
